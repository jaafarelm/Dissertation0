```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Load the images and labels
images_file = 'images.csv'
labels_file = 'labels.csv'

x = pd.read_csv(images_file, header=None).values
y = pd.read_csv(labels_file, header=None).values.flatten()

# Step 1: Verify dataset properties
print(f"Images shape: {x.shape}")
print(f"Labels shape: {y.shape}")
if x.shape[0] != len(y):
    raise ValueError("Mismatch between the number of images and labels!")

# Step 2: EDA - Visualize label distribution
plt.figure(figsize=(10, 6))
sns.countplot(x=y)
plt.title('Label Distribution')
plt.xlabel('Labels')
plt.ylabel('Frequency')
plt.show()

# Display sample images
unique_labels = np.unique(y)
plt.figure(figsize=(12, 12))
for idx, label in enumerate(unique_labels[:25]):  # Display up to 25 labels
    plt.subplot(5, 5, idx + 1)
    sample_index = np.where(y == label)[0][0]
    sample_image = x[sample_index].reshape(32, 32)
    plt.imshow(sample_image, cmap='gray')
    plt.title(f"Label: {label}")
    plt.axis('off')
plt.tight_layout()
plt.show()

# Step 3: Normalize images to [0, 1]
x = x / 255.0

# Step 4: Reshape images to 32x32x1 (grayscale)
x = x.reshape(-1, 32, 32, 1)
print(f"Data reshaped to: {x.shape}")

# Step 5: Split data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y
)
print(f"Training samples: {len(x_train)}, Test samples: {len(x_test)}")

# Step 6: Data Augmentation
datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=False,
    fill_mode='nearest'
)
datagen.fit(x_train)

# Step 7: Build the CNN model
num_classes = len(np.unique(y))  # Automatically detect the number of classes
print(f"Number of classes: {num_classes}")

model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 1)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(num_classes, activation='softmax')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Step 8: Define callbacks
callbacks = [
    tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True),
    tf.keras.callbacks.ModelCheckpoint('arabic_letters_model.keras', save_best_only=True)
]

# Step 9: Train the model
history = model.fit(
    datagen.flow(x_train, y_train, batch_size=32),
    epochs=50,
    validation_data=(x_test, y_test),
    callbacks=callbacks
)
print("Model training complete!")

# Step 10: Load the best model
model = tf.keras.models.load_model('arabic_letters_model.keras')
print("Best model loaded.")

# Step 11: Evaluate the model
loss, accuracy = model.evaluate(x_test, y_test)
print(f"Loss: {loss}")
print(f"Accuracy: {accuracy}")

# Step 12: Generate predictions and create confusion matrix
y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred_classes)
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()

# Classification report
print(classification_report(y_test, y_pred_classes))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.utils.class_weight import compute_class_weight
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Load the images and labels
images_file = 'images.csv'
labels_file = 'labels.csv'

x = pd.read_csv(images_file, header=None).values
y = pd.read_csv(labels_file, header=None).values.flatten()

# Step 1: Verify dataset properties
print(f"Images shape: {x.shape}")
print(f"Labels shape: {y.shape}")
if x.shape[0] != len(y):
    raise ValueError("Mismatch between the number of images and labels!")

# Step 2: EDA - Visualize label distribution
plt.figure(figsize=(10, 6))
sns.countplot(x=y)
plt.title('Label Distribution')
plt.xlabel('Labels')
plt.ylabel('Frequency')
plt.show()

# Display sample images
unique_labels = np.unique(y)
plt.figure(figsize=(12, 12))
for idx, label in enumerate(unique_labels[:25]):  # Display up to 25 labels
    plt.subplot(5, 5, idx + 1)
    sample_index = np.where(y == label)[0][0]
    sample_image = x[sample_index].reshape(32, 32)
    plt.imshow(sample_image, cmap='gray')
    plt.title(f"Label: {label}")
    plt.axis('off')
plt.tight_layout()
plt.show()

# Step 3: Normalize images to [0, 1]
x = x / 255.0

# Step 4: Reshape images to 32x32x1 (grayscale)
x = x.reshape(-1, 32, 32, 1)
print(f"Data reshaped to: {x.shape}")

# Step 5: Split data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y
)
print(f"Training samples: {len(x_train)}, Test samples: {len(x_test)}")

# Step 6: Data Augmentation (Enhanced)
datagen = ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=False,
    fill_mode='nearest'
)
datagen.fit(x_train)

# Step 7: Compute Class Weights
class_weights = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
class_weight_dict = dict(enumerate(class_weights))
print("Class weights computed successfully!")

# Step 8: Build the Enhanced CNN Model
num_classes = len(np.unique(y))  # Automatically detect the number of classes
print(f"Number of classes: {num_classes}")

model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), padding="same", input_shape=(32, 32, 1)),  # Added padding="same"
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Activation('relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),

    tf.keras.layers.Conv2D(64, (3, 3), padding="same"),  # Added padding="same"
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Activation('relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),

    tf.keras.layers.Conv2D(128, (3, 3), padding="same"),  # Added padding="same"
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Activation('relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),

    tf.keras.layers.Conv2D(256, (3, 3), padding="same"),  # Added padding="same"
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Activation('relu'),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(256),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Activation('relu'),
    tf.keras.layers.Dropout(0.6),
    tf.keras.layers.Dense(num_classes, activation='softmax')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Step 9: Define Callbacks (Enhanced Early Stopping and LR Scheduler)
lr_schedule = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=3,
    verbose=1,
    min_lr=1e-6
)

callbacks = [
    tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    tf.keras.callbacks.ModelCheckpoint('arabic_letters_model.keras', save_best_only=True),
    lr_schedule
]

# Step 10: Train the model
history = model.fit(
    datagen.flow(x_train, y_train, batch_size=32),
    epochs=50,
    validation_data=(x_test, y_test),
    callbacks=callbacks,
    class_weight=class_weight_dict
)
print("Model training complete!")

# Step 11: Load the best model
model = tf.keras.models.load_model('arabic_letters_model.keras')
print("Best model loaded.")

# Step 12: Evaluate the model
loss, accuracy = model.evaluate(x_test, y_test)
print(f"Loss: {loss}")
print(f"Accuracy: {accuracy}")

# Step 13: Generate predictions and create confusion matrix
y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred_classes)
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()

# Classification report
print(classification_report(y_test, y_pred_classes))
```
